```python
#Shakespeare monkey theorem: It states that, with an infinite amount of time, a monkey randomly typing in a typewriter 
# would recreate the famous writes works.

# Applying this theorem to python

# Using an inputed ssentenceee, we are going to randomly generate phrases with the same size as the inputed one, until
# one phrase matches with the inputed sentence, and to accelerate it we'll uuse genetic algorithms.
```

```python
#import modules

import random as ran
import numpy as np
from operator import attrgetter
```

```python
#define our phrase, using lower case letters.

phrase='the important thing in life is not tryumph, but the struggle'

# define caracteres that can be randomly selected to create strings.

caracteres=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z',' ',','];

```

```python
#Creating a class to deffine each individual, as a string and create and attribute for their fitness

class individual:
    
    def __init__(self,phrase,father = None ,mother= None):
        
        self.size=len(phrase)
        self.DNA=[]
        self.father=father
        self.mother=mother
        
        #Generate sentence
        if mother == None and father==None:
            for i in range(self.size):
                self.DNA.append(ran.choice(caracteres))
            self.DNA=''.join(self.DNA)
            
        else:
            p_mutation=0.04
            p_father=0.48
            p_mother=0.48
            mutation=ran.choice(caracteres)
            for i in range(self.size):
                genes=[self.father.DNA[i],self.mother.DNA[i],mutation]
                
                self.DNA.append(np.random.choice(genes, p=[p_father,p_mother,p_mutation]))
            
            self.DNA=''.join(self.DNA)
            
        index=0
        for i in range(self.size):
            if self.DNA[i]==phrase[i]:
                index+=1
            else:
                pass
            
        fit=index/self.size
        self.fit=fit
        
    
     # Print individual
    def __str__(self):
        return 'inidividual' + self.DNA
    
     #Same as attribute fit
    
    def fitness(self):
        counter=0
        for i in range(self.size):
            if self.DNA[i]==phrase[i]:
                counter+=1
            else:
                pass
        fit=counter/self.size
        return fit
    
 
```

```python
 #Generate population:
    
def population(size):
    p= []
    for i in range(size):
        
        p.append(individual(phrase))
        
    return p
```

```python
 #Selecting the individuals with greatest fitness
def selection(population):
    parents=[]
    population=sorted(population, key=attrgetter('fit'),reverse=True)
    
#This code sorts the population list using the fit attribute.
    if len(population)> 10:
            ten_percent=round(len(population)/10)
    else:
        ten_percent=1
        
    for i in range(ten_percent):
        parents.append(population[i])
        
    if len(parents)%2 ==0:
        return parents
    
    else:
        clone=parents[0]
        parents.append(clone)
        return parents
            
```

```python
#Reproduction using the selected individuals 
def next_generation(parents):
    population=[]
    fathers=[]
    mothers=[]
    for i in range(len(parents)):
        if i%2==0:
            fathers.append(parents[i])
        else:
            mothers.append(parents[i])
    while len(population) < (len(parents)*10):
        for a in range(len(fathers)):
            population.append(individual(phrase,fathers[a],mothers[a]))
    return population
```

```python
#Now the main function

#Putting time to the "natural selection"

import datetime
generation=0
p=population(1000)
Ti=datetime.datetime.now()

while True:
    p=sorted(p,key=attrgetter('fit'),reverse=True)
    print('\n')
    print('Generation: {}'.format(generation))
    for i in range(5):
        print(p[i],' ',p[i].fitness())
    
    if p[0].DNA == phrase:
        break
    
    progenitors=selection(p)
    p= next_generation(progenitors)
    print('\n')
    generation+=1
    
Tf=datetime.datetime.now()
delta_T=Tf-Ti
print('\n Evolution achieved !!! ')
print('\n It took {} seconds and {} generations, for evolution !'.format(delta_T.seconds,generation))

#The output will show the top 5 individuals in each generation, according to their fitness.
#After matching with the initial phrase, the output will tell how long and how many generations took to "Evolution".
# Evolution is the generated phrase being completely equal to the inputed phrase.
```
